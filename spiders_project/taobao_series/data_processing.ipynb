```python
a = [['自主访问', '1,193', '981', '554', '14.49%', '趋势'],
     ['购物车', '1,731', '1466', '846', '28.36%', '趋势'],
     ['我的淘宝', '950', '777', '634', '9.46%', '趋势'],
     ['直接访问', '0', '253', '37', '0.00%', '趋势'],
     ['付费流量', '994', '700', '333', '10.29%', '趋势'],
     ['淘宝客', '1533', '938', '327', '22.88%', '趋势'],
     ['品销宝- 明星店铺', '1010', '0', '605', '10.61%', '趋势'],
     ['直通车', '726', '700', '312', '5.65%', '趋势'],
     ['品销宝- 品牌首推', '0', '0', '0', '0.00%', '趋势'],
     ['超级推荐', '0', '405', '0', '0.00%', '趋势']]

df = pd.DataFrame(a)
df = df.drop(columns=[5])
# df.set_index([0], inplace=True)
df.columns = ['流量来源','本店交易指数', '竞店1交易指数', '竞店2交易指数', '本店支付金额']
df['data_time'] = ['2019/5/1', '2019/5/2'] * int(df.shape[0] / 2)

# df.columns = [['客群指数']*5,['本店交易指数', '竞店1交易指数', '竞店2交易指数', '本店支付金额','data_time']]

df['kinds'] = ['客群指数'] * df.shape[0]
# df.columns = [[]*,['本店交易指数', '竞店1交易指数', '竞店2交易指数', '本店支付金额']]
# df.index.name='流量来源'

# 更改数据类型
# df[['本店交易指数','竞店1交易指数','竞店2交易指数']] = df[['本店交易指数','竞店1交易指数','竞店2交易指数']].astype('int')
df
```

```python
df[['本店交易指数', '竞店1交易指数', '竞店2交易指数']] = df[['本店交易指数', '竞店1交易指数', '竞店2交易指数']].applymap(lambda x: int(x.replace(',','')))
df
```

```python
# 查看数据类型
print(df.dtypes)
```

```python
# 借助透视表进行数据重排,需要保证数据类型能够实现聚合函数，aggfun,默认取平均数
s = df.pivot_table(['本店交易指数','竞店1交易指数','竞店2交易指数'],index=['data_time','流量来源'], columns='kinds')
s
```

```python
s2 = df.pivot_table(['本店交易指数','竞店1交易指数','竞店2交易指数'],index=['data_time','流量来源'], columns='kinds',aggfunc=sum)
s2
```

```python
# 数据再分组，进行数据拼接
s1 = s['本店交易指数']
s2 = s['竞店1交易指数']
s3 = s['竞店2交易指数']
s_new = pd.concat([s1, s2, s3], keys=['本店交易指数','竞店1交易指数','竞店2交易指数'])
s_new
```
